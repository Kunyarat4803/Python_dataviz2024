#Quiz9
-วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่องทั้งหมดกี่เรื่อง  เรื่องอะไรบ้าง  
-มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
-คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อมากที่สุด

ข้อ1

````
# prompt: วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่องทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count = 0
list_name = []
for i, row in folk_df.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        count += 1
        list_name.append(row['title'])

print(f"มีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องทั้งหมด {count} เรื่อง")
print("รายชื่อนิทาน:")
for name in list_name:
    print(name)
````

ข้อ2

````
# prompt: วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่องทั้งหมดกี่เรื่อง เรื่องอะไรบ้างมีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

count_life = 0
count_local = 0
for i, row in folk_df.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        if row['type'] == 'นิทานชีวิต':
            count_life += 1
        elif row['type'] == 'นิทานประจำถิ่น':
            count_local += 1

if count_life > count_local:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานชีวิตมากกว่า")
elif count_local > count_life:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานประจำถิ่นมากกว่า")
else:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")
````

ข้อ3

````
# prompt:  คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อมากที่สุด

creator_counts = {}
for i, row in folk_df.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        creator = row['creator']
        if creator not in creator_counts:
            creator_counts[creator] = 0
        creator_counts[creator] += 1

max_count = 0
most_frequent_creator = ""
for creator, count in creator_counts.items():
    if count > max_count:
        max_count = count
        most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า 'แม่' หรือ 'พ่อ' มากที่สุดคือ {most_frequent_creator} ({max_count} เรื่อง)")
````
