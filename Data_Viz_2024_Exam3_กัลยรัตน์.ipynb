จุดที่2 ที่หน้าสนใจ คืออยากทราบพฤติกรรมที่ทำให้เกิดโรคมะเร็ง

```python
# prompt: พฤติกรรมที่ทำให้เกิดโรคมะเร็ง

# พฤติกรรมที่ทำให้เกิดโรคมะเร็ง
behavior_keywords = ['สูบบุหรี่', 'ดื่มเหล้า', 'กินอาหารปิ้งย่าง', 'นอนดึก', 'เครียด', 'อ้วน', 'ไม่ออกกำลังกาย', 'กินผักน้อย', 'ตากแดดจัด']

behavior_counts = {}
for keyword in behavior_keywords:
  behavior_counts[keyword] = Data_Docter_df['comments'].str.contains(keyword).sum()

sorted_behaviors = sorted(behavior_counts.items(), key=lambda x: x[1], reverse=True)

print("\nพฤติกรรมที่อาจทำให้เกิดโรคมะเร็งจากการกล่าวถึงมากไปน้อย:")
for behavior, count in sorted_behaviors:
  print(f"- {behavior}: {count} ครั้ง")
```

```python
import matplotlib.pyplot as plt

# ข้อมูลที่นับจำนวนการกล่าวถึงแต่ละพฤติกรรม
behavior_counts = {
    'สูบบุหรี่': 5,
    'ดื่มเหล้า': 3,
    'กินอาหารปิ้งย่าง': 4,
    'นอนดึก': 2,
    'เครียด': 3,
    'อ้วน': 1,
    'ไม่ออกกำลังกาย': 2,
    'กินผักน้อย': 3,
    'ตากแดดจัด': 0
}

# จัดเรียงพฤติกรรมตามจำนวนการกล่าวถึง
sorted_behaviors = sorted(behavior_counts.items(), key=lambda x: x[1], reverse=True)

# แยกประเภทพฤติกรรมและจำนวนการกล่าวถึง
behavior_types = [behavior for behavior, count in sorted_behaviors]
behavior_counts = [count for behavior, count in sorted_behaviors]

# สร้างกราฟแท่ง
plt.figure(figsize=(12, 8))
plt.bar(behavior_types, behavior_counts, color='lightcoral')
plt.xlabel("พฤติกรรมที่อาจทำให้เกิดโรคมะเร็ง")
plt.ylabel("จำนวนครั้งที่กล่าวถึง")
plt.title("จำนวนการกล่าวถึงพฤติกรรมที่ทำให้เกิดโรคมะเร็ง")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: หา comment คนที่สูบบุหรี่แล้วเป็นโรคมะเร็ง 3 อันดับแรก

# กรองข้อมูลเฉพาะโพสต์ที่กล่าวถึงทั้งการสูบบุหรี่และมะเร็ง
smoking_cancer_posts = Data_Docter_df[Data_Docter_df['comments'].str.contains('สูบบุหรี่') & Data_Docter_df['comments'].str.contains('มะเร็ง')]

# เรียงลำดับโพสต์ตามยอดไลค์จากมากไปน้อย
sorted_smoking_cancer_posts = smoking_cancer_posts.sort_values(by='ยอดไลค์', ascending=False)

# เลือก 3 อันดับแรก
top3_smoking_cancer_posts = sorted_smoking_cancer_posts.head(3)

# แสดงผลลัพธ์
print(top3_smoking_cancer_posts['comments'])
```
