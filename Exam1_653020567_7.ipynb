##ข้อ1 ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
# แยกปีจากคอลัมน์ month
df_Tidy ['year'] = df_Tidy ['month'].apply(lambda x: x.split()[1]) # ['July', '2018']
```

```python
# กรองข้อมูลสำหรับปี 2018 และ 2019
df_2018 = df_Tidy[df_Tidy['year'].str.contains('2018')]
df_2019= df_Tidy[df_Tidy['year'] .str.contains('2019')]
```

```python
# คำนวณค่าเฉลี่ย Chlo_a สำหรับแต่ละ station ระหว่างปี 2018 และปี 2019
mean_chlo_a_2018 = df_2018.groupby('station')['Chlo_a'].mean()
mean_chlo_a_2019 = df_2019.groupby('station')['Chlo_a'].mean()
```

```python
# สร้างตารางค่าเฉลี่ย Chlo_a ทุกๆ station ของแต่ละเดือนในปี 2018
mean_chlo_a_2018_by_month = df_2018.groupby(['station', 'month'])['Chlo_a'].mean().reset_index()
mean_chlo_a_2018_by_month
```

```python
# สร้างตารางค่าเฉลี่ย Chlo_a ทุกๆ station ของแต่ละเดือนในปี 2019
mean_chlo_a_2019_by_month = df_2019.groupby(['station', 'month'])['Chlo_a'].mean().reset_index()
mean_chlo_a_2019_by_month
```

```python
# คำนวณความแตกต่างของค่าเฉลี่ยระหว่างปี 2018 และ 2019
difference = mean_chlo_a_2019 - mean_chlo_a_2018
print(difference)
```

#ข้อ2สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

---



```python
import pandas as pd

# แทนที่ _ ด้วย ' ' ในคอลัมน์ month
df_Tidy['month'] = df_Tidy['month'].str.replace('_', ' ')

# แยกปีจากคอลัมน์ month
df_Tidy['year'] = df_Tidy['month'].apply(lambda x: x.split()[1]) # ['July', '2018']

# กรองข้อมูลสำหรับปี 2018 และ 2019
df_2018 = df_Tidy[df_Tidy['year'].str.contains('2018')]
df_2019 = df_Tidy[df_Tidy['year'].str.contains('2019')]

# รวมข้อมูลปี 2018 และ 2019
df_filtered = pd.concat([df_2018, df_2019])

# คำนวณค่าเฉลี่ยของ Chlo_b สำหรับแต่ละ station ในแต่ละเดือน
mean_chlo_b= df_filtered.groupby(['station', 'month'])['Chlo_b'].mean().reset_index()

# เปลี่ยนชื่อคอลัมน์ตามที่ต้องการ (ถ้าต้องการ)
mean_chlo_b.columns = ['Station', 'Month', 'Average_Chlo_b']

# แสดงผลลัพธ์
mean_chlo_b
```
